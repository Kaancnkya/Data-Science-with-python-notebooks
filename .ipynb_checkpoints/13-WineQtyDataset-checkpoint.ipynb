```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('13-WineQT.csv')
```

```python
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(13,13))

plt.subplot(2,2,1)
sns.set_style('whitegrid')
sns.barplot(x='quality', y='sulphates', data=df)
plt.title('quality by sulphates')
plt.xlabel('Quality')
plt.ylabel('sulphates')

plt.subplot(2,2,2)
sns.set_style('whitegrid')
sns.barplot(x='quality', y='volatile acidity', data=df)
plt.title('quality by volatile acidity')
plt.xlabel('Quality')
plt.ylabel('volatile acidity')

plt.subplot(2,2,3)
sns.set_style('whitegrid')
sns.boxplot(x='quality', y='sulphates', data=df)
plt.title('quality by sulphates')
plt.xlabel('Quality')
plt.ylabel('sulphates')

plt.subplot(2,2,4)
sns.set_style('whitegrid')
sns.boxplot(x='quality', y='volatile acidity', data=df)
plt.title('quality by volatile acidity')
plt.xlabel('Quality')
plt.ylabel('volatile acidity')

plt.tight_layout()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='volatile acidity', y='sulphates', data=df)
plt.title('volatile acidity and sulphates correlation')
plt.xlabel('volatile acidity')
plt.ylabel('sulphates')
plt.show()
```

```python
df_cleaned = df[(df['sulphates'] < 1.5) & (df['volatile acidity'] < 1.2)]
df_cleaned[['sulphates', 'volatile acidity']].corr()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='volatile acidity', y='sulphates', data=df_cleaned)
plt.title('volatile acidity and sulphates correlation')
plt.xlabel('volatile acidity')
plt.ylabel('sulphates')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
plt.subplot(1,2,1)
sns.set_style('whitegrid')
sns.boxplot(x='quality', y='sulphates', data=df_cleaned)
plt.title('quality by sulphates')
plt.xlabel('Quality')
plt.ylabel('sulphates')

plt.subplot(1,2,2)
sns.set_style('whitegrid')
sns.boxplot(x='quality', y='volatile acidity', data=df_cleaned)
plt.title('quality by volatile acidity')
plt.xlabel('Quality')
plt.ylabel('volatile acidity')

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df_cleaned, x='volatile acidity', y='sulphates', hue='quality', palette='coolwarm')
```

```python
sns.scatterplot(x="pH", y="fixed acidity", hue="quality", data=df)
```

```python
sns.scatterplot(x="alcohol", y="sulphates", hue="quality", data=df_cleaned)
```

```python
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
```
