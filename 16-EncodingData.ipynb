```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df[["sex", "class", "embark_town"]].isna().sum()
```

```python
df = df.dropna(subset=['embark_town'])
```

```python
df
```

```python
df[["sex", "class", "embark_town"]].isna().sum()
```

```python
#one hot encoding
```

```python
df['sex'].value_counts()
```

```python
df['class'].value_counts()
```

```python
df['embark_town'].value_counts()
```

```python
df_one_hot = pd.get_dummies(df, columns=['sex', 'embark_town'], drop_first=True)
```

```python
df_one_hot
```

```python
#Label encoder
```

```python
label_encoder = LabelEncoder()
df_label = df.copy()
```

```python
df_label
```

```python
df_label['sex'] = label_encoder.fit_transform(df_label['sex'])
```

```python
df_label
```

```python
#Ordinal Encoder
```

```python
df_ordinal = df.copy()
class_order = ['Third','Second','First']
```

```python
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal['class'] = ordinal_encoder.fit_transform(df_ordinal[['class']])
```

```python
df_ordinal
```

```python
fig, axes = plt.subplots (1,3, figsize=(15,5))
df['sex'].value_counts().plot(kind='bar', ax=axes[0], title='Original Categorial')
df_label['sex'].value_counts().plot(kind='bar', ax=axes[1], title='Original Categorial')
df_one_hot['sex_male'].value_counts().plot(kind='bar', ax=axes[2], title='Original Categorial')
```
