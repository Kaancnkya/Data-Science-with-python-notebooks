```python
def salary_category(salary):
    if salary < 50000:
        return "LOW"
    elif 80000>salary>50000:
        return "MEDIUM"
    else:
        return "HIGH"
```

```python
def adjust_performance(experience):
    if experience > 10:
        return 1
    else:
        return 0
        
```

```python
df['Adjusted'] = df['Experience'].apply(adjust_performance)
```

```python
df
```

```python
df['New_Score'] = df['Performance_Score'] + df['Adjusted']
```

```python
df
```

```python
def adjusted_score(row):
    if row['Experience'] > 10:
        return row['Performance_Score'] + 1
    else:
        return row['Performance_Score']
```

```python
df['Adjusted_Score'] = df.apply(adjusted_score, axis=1)
```
